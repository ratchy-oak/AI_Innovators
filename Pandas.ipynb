## 2.11 วิธีเลือกคอลัมน์ตามเงื่อนไขที่ต้องการ

Filter เฉพาะคอลัมน์ที่มีค่าตามที่เราต้องการโดยใช้ .loc ได้ โดยสามารถเลือก Filter แบบ .all() (ทุกค่าในคอลัมน์ต้องตรงตามเงื่อนไข) หรือ .any() (บางค่าในคอลัมน์ต้องตรงตามเงื่อนไข)

```python
import pandas as pd

# URL ของ dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# อ่าน dataset จาก URL
df_url = pd.read_csv(url)

# แสดงตัวอย่างข้อมูล
df_url.head(10)
```

```python
df_url.info()
```

```python
# เลือกเฉพาะคอลัมน์ที่เป็นตัวเลข
numeric_df = df_url.select_dtypes(include=[int, float])
print("DataFrame ที่มีคอลัมน์เป็นตัวเลข:")
print(numeric_df.head())
```

```python
# ใช้ .any() เพื่อเลือกคอลัมน์ที่มีค่าใดๆ มากกว่า 50
dataframe_any = numeric_df.loc[:, (numeric_df > 50).any()]
print("DataFrame ที่มีคอลัมน์ที่มีค่าใดๆ มากกว่า 50:")
print(dataframe_any.head())
```

```python
# ใช้ .all() เพื่อเลือกคอลัมน์ที่ทุกค่ามากกว่า 50
dataframe_all = numeric_df.loc[:, (numeric_df > 50).all()]
print("DataFrame ที่มีคอลัมน์ที่ทุกค่ามากกว่า 50:")
print(dataframe_all.head())
```

```python
# หาคอลัมน์ที่มี Missing Values
dataframe_with_missing = df_url.loc[:, df_url.isnull().any()]
print("DataFrame ที่มีคอลัมน์ที่มี Missing Values:")
print(dataframe_with_missing.columns)
```

```python
# หาคอลัมน์ที่ไม่มี Missing Values
dataframe_without_missing = df_url.loc[:, df_url.notnull().all()]
print("DataFrame ที่มีคอลัมน์ที่ไม่มี Missing Values:")
print(dataframe_without_missing.columns)
```

## 2.13 วิธีเพิ่มคอลัมน์ใหม่
สามารถเพิ่มคอลัมน์ใหม่ได้ 2 แบบ คือ

1. เพิ่มโดยอิงจากคอลัมน์เดิม (เช่น เอาคอลัมน์เดิม + 10 หรือ เอาคอลัมน์ A – คอลัมน์ B มีประโยชน์มากตอนทำ Feature Engineering)
2. เพิ่มคอลัมน์โดยตั้งค่า Fix ไปเลยสำหรับทุกแถว ส่วนใหญ่จะใช้วิธีนี้เวลาเราอยากได้ค่าอะไรแปลก ๆ ที่ต้องเขียนลูปเพื่อใส่ค่า ก็สร้างคอลัมน์แบบ Fix ค่าก่อน แล้วต่อด้วยลูป



```python
import pandas as pd

# URL ของ dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# อ่าน dataset จาก URL
df_url = pd.read_csv(url)
```

```python
# 1. เพิ่มคอลัมน์ใหม่โดยอิงจากคอลัมน์เดิม
df_url['Age_plus_10'] = df_url['Age'] + 10
df_url['Fare_minus_Age'] = df_url['Fare'] - df_url['Age']

# แสดงตัวอย่างข้อมูล
print(df_url[['Age', 'Fare', 'Age_plus_10', 'Fare_minus_Age']].head())
```

```python
# 2. เพิ่มคอลัมน์ใหม่โดยตั้งค่า Fix ไปเลยสำหรับทุกแถว
df_url['New_Column'] = 100

# แสดงตัวอย่างข้อมูล
print(df_url[['New_Column']].head())
```

## 2.15 การต่อ DataFrame (รวมข้อมูล DataFrame เช่น เอา CSV มาต่อกัน)

การต่อ Data Frame คือการเอา Data Set 2 ชุดมาต่อกันในแถวตั้งหรือแนวนอน สำหรับการต่อแบบปะติดไปเลย

มี 2 คำสั่งที่เหมือนกัน คือ concat กับ append แต่ให้ใช้ concat ไปเลย เพราะ append เป็นคำสั่งที่ไม่ Memory Efficient

```python
import pandas as pd

# สร้าง DataFrame ตัวอย่าง
df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

df2 = pd.DataFrame({
    'A': [7, 8, 9],
    'B': [10, 11, 12]
})

df3 = pd.DataFrame({
    'A': [17, 18, 19],
    'B': [20, 21, 22]
})
```

```python
df2
```

```python
# 2. ต่อ DataFrame ในแนวนอน
df_horizontal = pd.concat([df1, df2], axis=1)

print("DataFrame ที่ต่อในแนวนอน:")
print(df_horizontal)
```

```python
# 3. รวมมากกว่า 2 คอลัมน์ก็ได้
pd.concat([df1,df2,df3])
```

```python
# รวมเสร็จแล้ว reset index ให้ด้วย ควรใช้ ไม่งั้นจะเจอ row ID ซ้ำกันตอนรวมร่าง
pd.concat([df1,df2,df3], ignore_index=True)
```

## 2.20 รันคำสั่งที่เขียนเองกับทุกแถวใน 1 คอลัมน์

```python
import pandas as pd

# สร้าง DataFrame ตัวอย่าง
data = {
    'C1': [150, 200, 250, 300, 350],
    'C2': [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)
df.head()
```

```python
# ใช้ map และ lambda เพื่อลดค่าของแต่ละแถวในคอลัมน์ C1 ลง 100
df['C1'] = df['C1'].map(lambda x: x - 100)

# ใช้ map และ lambda เพื่อเพิ่มค่าของแต่ละแถวในคอลัมน์ C2 ขึ้น 3
df['C2'] = df['C2'].map(lambda x: x + 3)

print("DataFrame หลังจากปรับเปลี่ยนค่าของคอลัมน์ C1:")
print(df)
```

## 2.21 รันคำสั่งที่เขียนเองกับทุกค่า

```python
import pandas as pd

# สร้าง DataFrame ตัวอย่าง
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}
df = pd.DataFrame(data)
df.head()
```

## 2.22 คำนวณ Correlation & Covariance

**Correlation หรือ ค่าสหสัมพันธ์ **

เป็นการดูทิศทางความสัมพันธ์ระหว่างตัวแปร 2 ตัว โดยมี Correlation Coefficient (r) หรือ ค่าสัมประสิทธิ์สหสัมพันธ์ เป็นตัวบ่งชี้ถึงความสัมพันธ์นี้ ซึ่งค่าสัมประสิทธิ์สหสัมพันธ์นี้จะมีค่าอยู่ระหว่าง -1.0 ถึง +1.0

- หากมีค่าใกล้ -1.0 นั้นหมายความว่าตัวแปรทั้งสองตัวมีความสัมพันธ์กันอย่างมากในเชิงตรงกันข้าม

- หากมีค่าใกล้ +1.0 นั้นหมายความว่า ตัวแปรทั้งสองมีความสัมพันธ์กันโดยตรงอย่างมาก

- หากมีค่าเป็น 0 นั้นหมายความว่า ตัวแปรทั้งสองตัวไม่มีความสัมพันธ์ต่อกัน

Covariance หรือ ค่าความแปรปรวนร่วม คือ ค่าที่หาความแปรปรวนของข้อมูลสองชุด

ความแปรปรวนร่วม (Covariance) เป็นค่าชี้วัดการเปลี่ยนแปลงของค่าตัวแปรตัวใดตัวหนึ่งเมื่อมีการเปลี่ยนแปลงในค่าตัวแปรอีกตัวหนึ่งซึ่งแสดงระดับความสัมพันธ์เชิงเส้นตรงของตัวแปร 2 ตัวแปร

ค่าความแปรปรวมร่วม มีค่าที่เป็นไปได้อยู่ระหว่าง -oo และ oo
- ค่าความแปรปรวนร่วมที่มีค่ามากกว่า 0 แสดงถึงความสัมพันธ์ระหว่างตัวแปรทั้งสองโดยเมื่อค่าของตัวแปรหนึ่งมีค่ามาก ค่าของอีกตัวแปรหนึ่งมักมีค่ามากด้วย ในทางเดียวกันเมื่อค่าตัวแปรหนึ่งมีค่าน้อย ค่าของอีกตัวแปรหนึ่งมักมีค่าน้อยตามไปด้วย
- ค่าความแปรปรวนร่วมมีค่าน้อยกว่า 0 แสดงถึงความสัมพันธ์ระหว่างสองตัวแปรในทางตรงกันข้าม กล่าวคือเมื่อค่าของตัวแปรหนึ่งมีค่ามาก ค่าของอีกตัวแปรหนึ่งมักมีค่าน้อย ในทางกลับกันเมื่อค่าตัวแปรหนึ่งมีค่าน้อย ค่าของอีกตัวแปรหนึ่งมักมีค่ามาก
- ค่าความแปรปรวนร่วมที่มีค่าเท่ากับ 0 บ่งบอกถึงค่าของตัวแปรทั้งสองตัวแปรไม่มีความสัมพันธ์เชิงเส้นตรงต่อกัน

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# URL ของ dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# อ่าน dataset จาก URL
df_url = pd.read_csv(url)

# เลือกเฉพาะคอลัมน์ที่เป็นตัวเลข
numeric_df = df_url.select_dtypes(include=[int, float])

# คำนวณ Correlation
corr = numeric_df.corr()

# ตั้งค่า matplotlib figure
fig, ax = plt.subplots(figsize=(15, 8))

# วาด heatmap ของ Correlation Matrix โดยใช้ matplotlib
cax = ax.matshow(corr, cmap='coolwarm')

# เพิ่ม color bar
fig.colorbar(cax)

# ตั้งค่า ticks และ labels
ax.set_xticks(range(len(corr.columns)))
ax.set_yticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=90)
ax.set_yticklabels(corr.columns)

# เพิ่มค่า correlation บน heatmap
for (i, j), val in np.ndenumerate(corr.values):
    ax.text(j, i, f'{val:.2f}', ha='center', va='center', color='black')

# แสดงผลลัพธ์
plt.show()
```

## 2.23 คำนวณ Cross Tabulation

Cross Tabulation มีประโยชน์มากเวลาอยากรู้ว่ามี Data ที่ตรงกับกรุ๊ป A ของคอลัมน์ 1 และกรุ๊ป B ของคอลัมน์ 2 เท่าไหร่ เช่น มีนักเรียนผู้ชาย (คอลัมน์ gender) กี่คนในมัธยมปลาย (คอลัมน์ education)

```python
import pandas as pd

# URL ของ dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# อ่าน dataset จาก URL
df_url = pd.read_csv(url)

# ใช้ pandas.crosstab เพื่อสร้างตาราง cross-tabulation ระหว่าง 'Pclass' และ 'Survived'
aggregate = pd.crosstab(df_url.Pclass, df_url.Survived)

print("Cross-tabulation ระหว่าง 'Pclass' และ 'Survived':")
print(aggregate)
```

## 2.30 วิธีแทนค่า Missing Value ด้วยค่าเฉลี่ย (Mean Imputation)

วิธีหนึ่งในการแทนค่าที่หายไป คือการทำสิ่งที่เรียกว่า Mean Imputation หรือหาค่าเฉลี่ยของคอลัมน์นั้น แล้วเอามาแทนค่าที่หายไปนั่นเองครับ

ข้อดีของการทำ Mean Imputation คือ สามารถทำได้ง่าย แต่ก็ต้องระวังเรื่องข้อเสีย เช่น ทำแบบนี้จะเป็นการไม่สนใจความสัมพันธ์ระหว่างตัวแปร ทำให้เกิด Bias สูง ควรใช้เฉพาะเวลา Missing Value ไม่เยอะเท่านั้นครับ

สามารถรันโค้ดด้านล่างเพื่อทำ Mean Imputation ได้ง่าย ๆ เลย

```python
import numpy as np
import pandas as pd

# URL ของ dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# อ่าน dataset จาก URL
df_url = pd.read_csv(url)
df_url.info()
print(f'จำนวน missing value ก่อน: {df_url.Age.isna().sum()}') # Check if there are any missing values

meanAge = np.mean(df_url.Age) # Get mean value
print(f'อายุเฉลี่ย: {meanAge} ปี')
df.Age = df_url.Age.fillna(meanAge) # Fill missing values with mean
print(f'จำนวน missing value หลัง: {df.Age.isna().sum()}') # Check if there are any missing values
```
